# Import library

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# There I create dataframe
etudiants = {
'Name': ['Fatima ', 'Ahmed ', 'Youssef ', 'Khadija ', 'Amina '],
'Score_Math': [15 , 12 , 18 , 9 , 14] ,
'Score_Physics': [13 , 16 , 14 , 11 , 17]
}

df = pd.DataFrame(etudiants)
print(df)
```

```python
# I want know the mean of each learning
print("Mean score mathematics : ", df["Score_Math"].mean())
print("Mean score physics ' ", df["Score_Physics"].mean())
```

```python
# Now I visualize the result to each student of score mathematics

plt.figure(figsize=(6,6))
plt.bar(df["Name"], df["Score_Math"], color="blue")
plt.title("Score mathematics of each students")
plt.xlabel("Students")
plt.ylabel("Score mathematics")
plt.grid(axis='y')
plt.ylim(0,20)
```
