# Mise en situation : Agent immobilier

Vous êtes un agent immobilier à Casablanca. Un client vous appelle et vous dit :
" J'ai un appartement de 73 $m^{2}$. Combien est-ce que je peux le vendre ? "

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = {
    'Maison' : [1,2,3,4,5,6,7,8],
    'Surface' : [30, 45, 50, 55, 65, 80, 95, 110],
    'Prix' : [95, 130, 148, 155, 185, 220, 265, 305]
}

df = pd.DataFrame(data)

print("==== Notre datasets ====")
print(df)
print(f"\nDimension {df.shape[0]} lignes x {df.shape[1]} colonnes")
```

```python
plt.figure(figsize=(8,8))
plt.scatter(df["Surface"], df["Prix"], color='blue', linewidth=5, label='Prix par rapport au surface')
plt.axvline(x=73, color='red', linestyle='--', label='73 $m^{2}$')

plt.title("Prix de maison en fonction des surfaces", fontsize=12, fontweight='bold')
plt.xlabel("Surface ($m^{2}$)", fontsize=10)
plt.ylabel("Prix (€)", fontsize=10)
plt.grid(True)
plt.legend()
plt.show()
```
