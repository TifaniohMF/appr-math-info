# Si on prend y = 17 + 2,6x comme modèle

```python
x_vals = np.linspace(0,150,1000)
y_vals = f(x_vals)

plt.figure(figsize=(8,6))
plt.scatter(df["Surface"], df["Prix"], color='blue', linewidth=5, label='Prix par rapport au surface')
plt.plot(x_vals, y_vals, color='green', label='y = 17 + 2.6x')

plt.title("Prix de maison en fonction des surfaces", fontsize=12, fontweight='bold')
plt.xlabel("Surface ($m^{2}$)", fontsize=10)
plt.ylabel("Prix (€)", fontsize=10)
plt.grid(True)
plt.legend()
plt.show()
```
